```python
import pandas as pd
d=pd.read_csv("/content/Japan_life_expectancy.csv")
d.head()

```

```python
x = d.drop(columns=['Welfare_exp', 'Prefecture'])
print(x)
y=d['Welfare_exp']
print(y)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pp=model.predict( X_train)
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
mse2=mean_squared_error(y_train,y_pp)
print("Mean Squared Error:", mse)
print(mse2)
```

```python
from sklearn.linear_model import Ridge
alpha = 0.01
ridge_model = Ridge(alpha=alpha)
ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
y_pp = ridge_model.predict(X_train)
mse2 = mean_squared_error(y_train, y_pp)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
print("Mean Squared Error:", mse2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values for Ridge Regression")
plt.show()
```

```python
from sklearn.linear_model import Lasso
alpha =0.3
lasso_model = Lasso(alpha=alpha)
lasso_model.fit(X_train, y_train)
y_pred = lasso_model.predict(X_test)
y_pp = lasso_model.predict(X_train)
mse = mean_squared_error(y_test, y_pred)
mse1= mean_squared_error(y_train, y_pp)
print("Mean Squared Error:", mse)
print("Mean Squared Error:", mse1)
```
